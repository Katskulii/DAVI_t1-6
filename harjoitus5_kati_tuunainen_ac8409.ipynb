## 1. Säädatan esittäminen
Esitetään säädataa, joka on ladattu Ilmatieteen laitoksen WFS rajapintapalvelusta. Ennen säädatan esittämistä se tulee käsitellä seuraavasti:

- lataa data DataFrameen url:sta 'https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta5/harjoitukset/saatiedot.csv' ottaen huomioon sarakkeiden erotinmerkin ',' ja desimaalierottimen '.',
- datan ensimmäisessä sarakkeessa (tai indeksissä) on esitetty aika muodossa vuosi-kuukausi-päivä tunnit:minuutit:sekunnit,
- muodosta em. aika sarakkeen kaikkia kenttiä (päivämäärä ja kellonkaika) hyödyntämällä uuteen DataFrameen sarake 'aika, s', joka kertoo kuinka paljon aikaa on kulunut sekunteina ensimmäisestä mittaushavainnosta eteenpäin,

- tulosta tekemästäsi em. DataFramesta viisi ensimmäistä riviä.

Kun olet käsitellyt datan, tee xy - tasolle seuraavanlainen kuvaaja:

- aika sarakkeesta 'aika, s' on esitetty vaaka-akselilla ja ilman lämpötila sarakkeesta 'Air temperature, degC' on esitetty pystyakselilla,
- em. kuvaajassa datapisteitä on merkitty punaisella pisteellä,
- punaista pistettä esittävälle datalle on merkitty selite (legend), jossa lukee 'ilman lämpötila',
- vaaka-akelin otsikoksi on kirjoitettu 'aika, s' ja pystyakselin otsikoksi 'ilman lämpötila 
o

 ',
- huomaa, että akselin otsikkoon voit merkitä fysikaalisen yksikön TeX syntaksilla,
- lisää kuvaajaan myös ruudukko, jossa on musta katkoviiva.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta5/harjoitukset/saatiedot.csv"

df = pd.read_csv(url, sep=',', decimal='.')

df['Unnamed: 0'] = pd.to_datetime(df['Unnamed: 0'])

new_df = pd.DataFrame(df['Air temperature, degC'])

time_diff = df['Unnamed: 0'] - df['Unnamed: 0'][0]
time_diff_sec = time_diff.dt.total_seconds()

new_df['aika, s'] = time_diff_sec

print(new_df.head())

plt.plot(new_df['aika, s'], new_df['Air temperature, degC'], 'ro', markersize=4, label='ilman lämpötila')
plt.xlabel('aika, s')
plt.ylabel('ilman lämpötila $^{o}C$')
plt.legend()
plt.grid(linestyle='--', color='black')
plt.show()
```

Kun olet käsitellyt datan, tee xy - tasolle seuraavanlainen kuvaaja:

- aika sarakkeesta 'aika, s' on esitetty vaaka-akselilla ja ilman lämpötila sarakkeesta 'Air temperature, degC' on esitetty vasemman puoleisella pystyakselilla ja tuulen nopeus sarakkeesta 'Wind speed, m/s' on esitetty oikean puoleisella pystyakselilla,

- aseta kuvan otsikoksi 'Säätila Jyväskylän lentoasemalla'.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta5/harjoitukset/saatiedot.csv"

df = pd.read_csv(url, sep=',', decimal='.')

df['Unnamed: 0'] = pd.to_datetime(df['Unnamed: 0'])
new_df = pd.DataFrame(df['Air temperature, degC'])
time_diff = df['Unnamed: 0'] - df['Unnamed: 0'][0]
time_diff_sec = time_diff.dt.total_seconds()
wind_df = pd.DataFrame(df['Wind speed, m/s'])
new_df = pd.concat([new_df, wind_df], axis=1)
new_df['aika, s'] = time_diff_sec

print(new_df.head())

fig, ax1 = plt.subplots()
ax1.set_facecolor('#D7F8FF')

ax1.plot(new_df['aika, s'] / 1000, new_df['Air temperature, degC'],'ro', markersize=4, label='ilman lämpötila')
ax1.set_xlabel('aika, $^{.}10^3$ s')
ax1.set_ylabel('ilman lämpötila $^{o}C$')

plt.title('Säätila Jyväskylän lentoasemalla')
plt.grid(linestyle='--', color='black')

ax2 = ax1.twinx()
ax2.plot(new_df['aika, s'] / 1000, new_df['Wind speed, m/s'], 'bo', markersize=4, label='tuulen nopeus')
ax2.set_ylabel('tuulen nopeus m/s')

ax2.legend(loc='upper left')
ax1.legend(loc='lower right')

plt.show()
```

## 3. Valioliigan sarjataulukko (2)
Esitetään valioliigadataa:

- lataa data DataFrameen url:sta 'url = "https://student.labranet.jamk.fi/~pelju/data_av/samples/premium_league_2018_19/england-premier-league-teams-2018-to-2019-stats.csv"

Esitetään valittujen valioliigajoukkueiden koti- ja vierasvoitot samassa pylväsdiagrammissa. Ennen datan esittämistä se pitää käsitellä seuraavasti:

- lataa data DataFrameen tehtävässä annetusta url:sta
- lisää uuteen DataFrameen sarakkeet team_name, common_name, season, wins, draws ja losses.
- valitse DataFrameen seuraavat joukkueet (common_name): 'Arsenal', 'Tottenham Hotspur', 'Manchester City', 'Manchester United', 'Chelsea', 'Liverpool', 'Everton'
- Lyhennä joukkueiden nimet kuvaajaa varten tekemällä uusi sarake short_name: 'Arsenal', 'Tottenham', 'ManCity', 'ManU', 'Chelsea', 'Liverpool', 'Everton'
- Lisää dataframeen kokonaan uusi sarake points (pisteet), jonka laskukaava on voitot * 3 + tasapelit (tappioista saa nolla pistettä): eli 'wins' * 3 + 'draws'

Kun olet käsitellyt datan, tee (x,y)-tasolle käännetty pylväsdiagrammi, jossa y-akselilla on joukkueen lyhennetty nimi ja x-akselilla on joukkueen keräämät pisteet:

- pyri kääntämään y-akselin otsikoita 30 astetta (vinkki: plt.yticks())
- aseta y-akselin pylväille myös otsikko, johon haet tiedot kaudesta (season) ja pelien määrästä (matches_played) alkuperäisestä DataFramesta
- lisää kuvaan myös ruudukko (vinkki: grid)

```python
url = "https://student.labranet.jamk.fi/~pelju/data_av/samples/premium_league_2018_19/england-premier-league-teams-2018-to-2019-stats.csv"

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(url)
sel_df = df.loc[:,['team_name', 'common_name', 'season', 'wins', 'draws', 'losses']]
teams = ['Arsenal', 'Tottenham Hotspur', 'Manchester City', 'Manchester United', 'Chelsea', 'Liverpool', 'Everton']
mask_df = sel_df['common_name'].isin(teams)
new_df = sel_df[mask_df].copy()
new_df['short_name'] = new_df['common_name'].replace('Tottenham Hotspur', 'Tottenham')
new_df['short_name'] = new_df['short_name'].replace('Manchester City', 'ManCity')
new_df['short_name'] = new_df['short_name'].replace('Manchester United', 'ManU')
new_df['points'] = new_df['wins']*3 + new_df['draws']
new_df = new_df.sort_values('points', ascending=True)

new_df

ax = sns.barplot(x='points', y='short_name', data = new_df, order = new_df.sort_values('points', ascending=False).short_name, edgecolor='black')
ax.set_xlabel('Team name')
ax.set_title(f'England: {df["season"].unique()[0]} , Matches played: {df["matches_played"][0]}')
plt.ylabel(None)
plt.yticks(rotation=30)
plt.grid(linestyle='--', color='black')

plt.show()
```

## 4. Autodatan esittäminen
### Datan esikäsittely
Lataa data DataFrameen URL-osoitteesta: https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta5/harjoitukset/autot.csv, kun olet huomioinut sarakkeiden erotinmerkit '\t'. Tallenna uuteen DataFrameen vain ne rivit, joissa autonvalmistajalla (sarake Mh) on enemmän kuin 50 autoa. Muuta uudessa DataFramessa sarakkeen Ft merkkijonojen 'DIESEL' tilalle 'Diesel' ja 'PETROL' tilalle 'Petrol'. Jos sarakkeessa Ft on merkkijono 'NG', korvaa se merkkijonolla 'Natural gas*'.

### Datan visualisointi
Tee esimerkkikuvan kaltainen kuvaaja, jossa hyödynnät tuottamasi DataFramen sarakkeita autojen polttoainetyypin Ft ja auton valmistajan Mh osalta.

```python
url = "https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta5/harjoitukset/autot.csv"

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(url, sep='\t')
df_filtered = df.groupby('Mh').filter(lambda x: len(x) > 50)
df_filtered['Ft'] = df_filtered['Ft'].replace('DIESEL', 'Diesel')
df_filtered['Ft'] = df_filtered['Ft'].replace('PETROL', 'Petrol')
df_filtered['Ft'] = df_filtered['Ft'].replace('NG', 'Natural gas')

df_filtered

ax = sns.countplot(y='Mh', hue='Ft', data=df_filtered, edgecolor='black')
ax.legend(title='Fuel type', loc='lower right')
ax.set_xlabel('Number of cars')
ax.set_ylabel('Manufacturer name')
ax.set_title('Number of cars by fuel type')
plt.grid(linestyle='--', color='black')
ax.set_axisbelow(True)

plt.show()
```

Tee esimerkkikuvan kaltainen histogrammi, josta käy ilmi

```python
url = "https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta5/harjoitukset/autot.csv"

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(url, sep='\t')
df_filtered = df.groupby('Mh').filter(lambda x: len(x) > 50)
df_filtered = df_filtered.loc[df_filtered['Enedc (g/km)'] > 0]
df_filtered

fig, ax = plt.subplots(figsize=(12, 6))

plt.hist(df_filtered['Enedc (g/km)'], bins=100, log=True, density=True, color='lightblue', label='Specific $CO_2$ Emissions')
plt.hist(df_filtered['Enedc (g/km)'], bins=100, cumulative=True, density=True, histtype='step', color='orange', label='Cum.specific $CO_2$ Emissions')

ax.set_xlabel('Specific $CO_2$ Emissions, (g/km)')
ax.set_ylabel('Probability density')
ax.set_yscale('log')
ax.legend(loc='lower right')

plt.show()
```
